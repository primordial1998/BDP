## Question 12

```python
df_child = df[df.primary_type == 'OFFENSE INVOLVING CHILDREN']
fig, ax = plt.subplots(figsize=(11,7))
df_child.sort_values('community_area', inplace=True)
df_child.plot(x='community_area', y='count', kind='bar', ax=ax, color='tab:blue')
ax.set_xticks(ax.get_xticks()[::2])
plt.show()
```
